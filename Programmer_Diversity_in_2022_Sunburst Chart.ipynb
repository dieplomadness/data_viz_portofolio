```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
import plotly.offline as pyo
```

```python
df = pd.read_csv('/Users/aryasusila/Documents/33 PROGRAMMER DIVERSITY/DATA PROGRAMMER SURVEY/Sorted_Survey.csv', index_col='Respondent')
```

```python
fig = px.sunburst(df, path=["Gender", "Trans", "Sexuality"], color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textinfo='label+percent parent', textfont_size=18)
fig.update_layout(font_family='verdana')
pyo.plot(fig, filename='Programmer Diversity in 2022_Sunburst Chart.html')
```
